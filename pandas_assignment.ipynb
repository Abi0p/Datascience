```python
import pandas as pd

data_products = {
    'ProductID': ['P1001', 'P1002', 'P1003', 'P1004', 'P1005', 'P1006', 'P1007', 'P1008', 'P1009', 'P1010',
                  'P1011', 'P1012', 'P1013', 'P1014', 'P1015'],
    'ProductName': ['Widget', 'Gadget', 'WidgetPro', 'Gizmo', 'Doodad', 'Thingy', 'Doohickey', 'Gimmick', 'Contraption', 'Device',
                    'Mechanism', 'Apparatus', 'Implement', 'Tool', 'Instrument'],
    'Price': [25.50, 45.75, 55.00, 12.95, 8.50, 22.50, 3.75, 7.80, 10.00, 33.50,
              45.00, 12.00, 14.95, 21.75, 17.50]
}
products = pd.DataFrame(data_products)
```

```python
data_sales = {
    'SaleID': list(range(101, 116)),
    'ProductID': ['P1001', 'P1003', 'P1002', 'P1005', 'P1001', 'P1006', 'P1007', 'P1008', 'P1009', 'P1004',
                  'P1010', 'P1011', 'P1013', 'P1014', 'P1015'],
    'Quantity': [3, 1, 2, 5, 10, 4, 2, 8, 1, 3, 7, 6, 3, 1, 2],
    'CustomerID': ['C001', 'C003', 'C002', 'C002', 'C001', 'C004', 'C005', 'C003', 'C005', 'C001',
                   'C002', 'C003', 'C004', 'C005', 'C001']
}
sales = pd.DataFrame(data_sales)
```

```python
data_customers = {
    'CustomerID': ['C001', 'C002', 'C003', 'C004', 'C005', 'C006', 'C007', 'C008', 'C009', 'C010',
                   'C011', 'C012', 'C013', 'C014', 'C015'],
    'CustomerName': ['Alpha Corp', 'Beta LLC', 'Gamma Inc', 'Delta Co', 'Epsilon Ltd', 'Zeta Inc', 'Eta LLC', 'Theta Corp', 'Iota PLC', 'Kappa LLC',
                     'Lambda Co', 'Mu Inc', 'Nu Ltd', 'Xi Corp', 'Omicron PLC'],
    'Location': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose',
                 'Austin', 'Jacksonville', 'Fort Worth', 'Columbus', 'Charlotte']
}
customers = pd.DataFrame(data_customers)
```

# **Tasks**

1. **What is the total revenue generated by each customer?**


```python
import pandas as pd

merged_data = pd.merge(sales, products, on='ProductID')
print (merged_data)
merged_data['Revenue'] = merged_data['Quantity'] * merged_data['Price']
customer_revenue = merged_data.groupby('CustomerID')['Revenue'].sum().reset_index()
customer_revenue = pd.merge(customer_revenue, customers[['CustomerID', 'CustomerName']], on='CustomerID')
print(customer_revenue)
```

2. **Which product has the highest total sales revenue?**


```python
import pandas as pd
merged_data = pd.merge(sales, products, on='ProductID')
merged_data['Revenue'] = merged_data['Quantity'] * merged_data['Price']
product_revenue = merged_data.groupby('ProductID')['Revenue'].sum().reset_index()
highest_revenue_product = product_revenue.loc[product_revenue['Revenue'].idxmax()]
highest_revenue_product = pd.merge(highest_revenue_product.to_frame().T, products[['ProductID', 'ProductName']], on='ProductID')
print(highest_revenue_product)
```

3. **How many unique customers bought each product?**


```python
import pandas as pd
merged_data = pd.merge(sales,customers,on='CustomerID')
product_customer_count = merged_data.groupby('ProductID')['CustomerID'].nunique().reset_index()
product_customer_count = pd.merge(product_customer_count, products[['ProductID', 'ProductName']], on='ProductID')
print(product_customer_count)

```

4. **Which customer bought the most units of any single product in a transaction?**

```python

max_units_transaction = sales.sort_values(by='Quantity', ascending=False).head(1)
max_units_transaction = pd.merge(
    max_units_transaction,
    customers[['CustomerID', 'CustomerName']],
    on='CustomerID'
)
max_units_transaction = pd.merge(
    max_units_transaction,
    products[['ProductID', 'ProductName']],
    on='ProductID'
)

print(max_units_transaction[['CustomerID', 'CustomerName', 'ProductID', 'ProductName', 'Quantity']])
```

5. **What is the average price of the products purchased by each customer?**

```python
merged_data = pd.merge(sales,products,on='ProductID')
customer_avg_price = merged_data.groupby('CustomerID')['Price'].mean().reset_index()
customer_avg_price=pd.merge(customer_avg_price,customers[['CustomerID','CustomerName']],on='CustomerID')
print(customer_avg_price)
```

6. **List all products bought by customers located in 'New York'.**

```python
merged_data=pd.merge(sales,customers,on='CustomerID')
new_york_customers = merged_data[merged_data['Location']=='New York']
new_york_products = pd.merge(new_york_customers,products,on='ProductID')
print(new_york_products[['CustomerName','ProductName']])

```

7. **Which location had the highest number of distinct products sold?**

```python

sales_with_location = pd.merge(sales, customers[['CustomerID', 'Location']], on='CustomerID')
distinct_products_per_location = sales_with_location.groupby('Location')['ProductID'].nunique().reset_index()
location_with_max_distinct_products = distinct_products_per_location.loc[distinct_products_per_location['ProductID'].idxmax()]

print(location_with_max_distinct_products)
```

8. **What is the total quantity of products sold in 'Los Angeles'?**

```python
quantity_with_location=pd.merge(sales,customers,on ='CustomerID')
quantity_with_location=quantity_with_location[quantity_with_location['Location']=='Los Angeles']
total_quantity=quantity_with_location['Quantity'].sum()
print(total_quantity)
```

9. **Identify the customer who has spent the most money on a single transaction.**

```python

sales_with_price = pd.merge(sales, products[['ProductID', 'Price']], on='ProductID')
sales_with_price['TotalAmount'] = sales_with_price['Quantity'] * sales_with_price['Price']
max_spent_transaction = sales_with_price.loc[sales_with_price['TotalAmount'].idxmax()]
max_spent_transaction = pd.merge(
    max_spent_transaction.to_frame().T,
    customers[['CustomerID', 'CustomerName']],
    on='CustomerID'
)
print(max_spent_transaction[['CustomerID', 'CustomerName', 'ProductID', 'Quantity', 'Price', 'TotalAmount']])
```

10. **Which product is the most popular in terms of quantity sold?**

```python

total_quantity_per_product = sales.groupby('ProductID')['Quantity'].sum().reset_index()
most_popular_product = total_quantity_per_product.loc[total_quantity_per_product['Quantity'].idxmax()]
most_popular_product = pd.merge(most_popular_product.to_frame().T, products[['ProductID', 'ProductName']], on='ProductID')
print(most_popular_product[['ProductID', 'ProductName', 'Quantity']])
```

11. **How many transactions have occurred in each location?**

```python
sales_with_location = pd.merge(sales, customers[['CustomerID', 'Location']], on='CustomerID')
transactions_per_location = sales_with_location.groupby('Location').size().reset_index(name='TransactionCount')
print(transactions_per_location)
```

12. **What is the average number of items per transaction for each location?**

```python

sales_with_location = pd.merge(sales, customers[['CustomerID', 'Location']], on='CustomerID')
location_stats = sales_with_location.groupby('Location').agg(
    TotalItems=('Quantity', 'sum'),
    Transactions=('Quantity', 'size')
).reset_index()
location_stats['AvgItemsPerTransaction'] = location_stats['TotalItems'] / location_stats['Transactions']
print(location_stats[['Location', 'AvgItemsPerTransaction']])
```

13. **Which customer has the widest variety of products purchased?**

```python

sales_with_products = pd.merge(sales, products[['ProductID', 'ProductName']], on='ProductID')
unique_products_per_customer = sales_with_products.groupby('CustomerID')['ProductID'].nunique().reset_index()
customer_with_widest_variety_index = unique_products_per_customer['ProductID'].idxmax()
customer_with_widest_variety = unique_products_per_customer.loc[[customer_with_widest_variety_index]]

# Step 4: Merge with customers to get the customer name and other details
# Instead of directly assigning, use a new variable or keep the original:
merged_customer_data = pd.merge(customer_with_widest_variety, customers[['CustomerID', 'CustomerName']], on='CustomerID')


# Display the result
print(merged_customer_data[['CustomerID', 'CustomerName', 'ProductID']])
```
